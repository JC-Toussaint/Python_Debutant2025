# Le jeu de __ta__ vie
Le Jeu de la Vie est un automate cellulaire conçu par John Horton Conway en 1970. Il s'agit d'une simulation mathématique où aucune intervention du joueur n'est nécessaire pendant son déroulement. Ce jeu est fondé sur un modèle d'automate cellulaire où chaque état conduit automatiquement à l'état suivant en suivant des règles prédéfinies.

Le jeu se déroule sur une grille à une dimension, théoriquement infinie dans sa version la plus simpliste. Les cases de cette grille, communément appelées "cellules" par analogie avec les cellules vivantes, peuvent être dans l'un des deux états distincts : "vivante" ou "morte".

En dimension 1, à chaque itération, l'état d’une cellule est entièrement déterminée par l’état de ses deux cellules premières voisines, selon les règles suivantes, comme le suggère la figure ci-dessous:

![Alt text](game_of_life1D_rule126.png)

* une cellule vivante possédant deux cellules voisines vivantes meurt
* une cellule morte possédant deux cellules voisines mortes reste morte
* sinon quel que soit son voisinage et son état, elle passe à l'état vivant.

* Une cellule vivante sera représentée par un carré noire dont le code unicode est :
bsc='\u25a0'  # black square character
* Une cellule morte sera représentée par un carré blanc dont le code unicode est :
wsc='\u25a1'  # white square character
